<h2><span style="color:seagreen">Concepto:</span> </h2>
<hr style="border:0.5px solid seagreen">
La regresión lineal es un modelo estdístico que permite aproximar las relaciones entre una variable dependiente $Y$ y un conjunto de variables independientes $X_{i}$. Este es uno de los múltiples modelos de regresión que existen que permiten aproximar y predecir comportamientos a cierto tipos de datos. La regresión lineal aproxima la relación a los datos a una recta en $m$ dimensiones, donde $m$ corresponde a la cantidad de variables independientes del sistema. El modelo puede presentarse de la siguiente manera: <br>
<center>$Y=\beta_{0}+\beta_{1}X_{1}+\beta_{2}X_{2}+...+\beta_{m}X_{m}+\epsilon$</center>
Donde $\beta_{i}$ son coeficientes relacionados con la pendiente de cada variable dependiente $X_{i}$ y $\epsilon$ es el error o variable aleatoria.

<h2><span style="color:seagreen">Ejemplo:</span> </h2>
<hr style="border:0.5px solid seagreen">
Se sabe que las resistencias son un componente pasivo lineal, es decir que su comportamiento se aproxima a una recta cuando están expuestas a un voltaje o una cortiente siguiendo la ley de ohm. 
<center>$I = \frac{V}{R}$</center>
Para este ejemplo tomaremos como variable dependiente a la corriente ($I$) y como variable independiente el voltaje ($V$)
A continuación se presenta una tabla con datos tomados experimentalmente para una resistencia de $680 \Omega$

```python
import pandas as pd

data = pd.read_excel("Resistencia.xlsx")
data
```

```python
import matplotlib.pyplot as plt


x = data["Voltaje (V)"]
y = data["Corriente (mA)"]
plot = plt.scatter(x, y)
plt.ylabel("Corriente (mA)")
plt.xlabel("Voltaje (V)")
plt.show()
```

Como se puede ver en la tabla y la Gráfica los datos se acercan a lo que es una linea recta con una pendiente específica. Para hallar la regresión existen múltiples métodos, como por ejemplo el método de mínimos cuadrados.
Para este caso la ecuación que se obtiene es:
<center>$Y = \frac{1}{680}X$</center>

```python
import numpy as np
fig, ax = plt.subplots()
ax.scatter(x, y)
b, a = np.polyfit(x, y, deg=1)
xseq = np.linspace(0, 18, num=100)
ax.plot(xseq, xseq*(1000/680), color="k", linestyle="--", label="y=x/680")
plt.ylabel("Corriente (mA)")
plt.xlabel("Voltaje (V)")
plt.legend()
plt.show()
```

La ecuación ecuación supuesta es adecuada la linea punteada pasa muy cerca a los valores de los datos tomados experimentalmente, es decir nuestra aproximación es correcta.
